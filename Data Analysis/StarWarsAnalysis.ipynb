```python
import pandas as pd

star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
star_wars.head(10)
```

```python
star_wars.columns
```

```python
# Removing RespondantID's that are NaN
condition = pd.notnull(star_wars['RespondentID'])
star_wars = star_wars[condition]
```

```python
#Use this to see unique values in column
column = star_wars['Have you seen any of the 6 films in the Star Wars franchise?']
print (column.value_counts())
```

```python
# Using the map function to convert. Creating a dictionary first. 
# Then using the map function to convert the answers

yes_no = {
    "Yes": True,
    "No": False
}

for col in ['Have you seen any of the 6 films in the Star Wars franchise?',
           'Do you consider yourself to be a fan of the Star Wars film franchise?']:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

```python
#Changing values of columns for which movies the respondant has seen.

import numpy as np

true_false = {
    'Star Wars: Episode I  The Phantom Menace': True,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True,
    np.nan: False
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(true_false)
```

```python
# Renaming columns to something more descrptive
star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": 'seen_2',
    "Unnamed: 5": 'seen_3',
    "Unnamed: 6": 'seen_4',
    "Unnamed: 7": 'seen_5',
    "Unnamed: 8": 'seen_6'
    })
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'ranking_1',
    'Unnamed: 10': 'ranking_2',
    'Unnamed: 11': 'ranking_3',
    'Unnamed: 12': 'ranking_4',
    'Unnamed: 13': 'ranking_5',
    'Unnamed: 14': 'ranking_6'
    })
```

```python
# Determining Average ranking for each of the 6 movies
import matplotlib.pyplot as plt
%matplotlib inline

print (star_wars[star_wars.columns[9:15]].mean())

#Range and Height arguements need to match
plt.bar(range(6), height = star_wars[star_wars.columns[9:15]].mean())

plt.show()
```

```python
# Determining which movie was seen the most
print(star_wars[star_wars.columns[3:9]].sum())
plt.bar(range(6), height = star_wars[star_wars.columns[9:15]].mean())
plt.show()
```

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

# Male/Female Analysis for most seen movie
# print (males[males.columns[3:9]].sum())

fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(121)
plt.bar(range(6), males[males.columns[3:9]].sum())
plt.title("Most Seen for Males")
ax2 = fig.add_subplot(122)
plt.bar(range(6), females[females.columns[3:9]].sum())
plt.title("Most Seen for Females")


plt.show()
#Highest watched movie is Episode V for males
# print (males[males.columns[9:15]].mean())
fig2 = plt.figure(figsize=(8,4))
ax1 = fig2.add_subplot(121)
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.title("Movies Ranked By Males")
ax2 = fig2.add_subplot(122)
plt.bar(range(6), females[females.columns[9:15]].mean())
plt.title("Movies Ranked By Females")

plt.show()
```
